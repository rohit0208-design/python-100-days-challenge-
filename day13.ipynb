```python
# variable length of postitinal arguements
def add(a,*args):
    print(a)
    print(args)
add(1,2,3,4,5,6,7,8)
```

```python
# variable length of postitinal arguements
def add(a,*args):
    print(a)
    print(args)
    
    s=0
    for num in args:
        s+=num 
        return s
print(add(1,2,3,4,5))
```

```python
#convert list of number into even and odd respective lists 
l = list(map(int,input().split()))
even = []
odd = []
for num in l:
    if num  %2==0:
       even.append(num)
    else:
        odd.append(num)

print("even list ",even)
print("odd list ",odd)
```

```python
# marks analyzer 
# in list contains students marks , find the pass percentage and fail percentage 35%
def student_marks(num):
    if num <=35:
        return "fail"
    else:
        return "pass"

number = int(input())
res =student_marks(number)
print(res)

```
