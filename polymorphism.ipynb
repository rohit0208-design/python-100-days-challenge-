### polymorphism

```python
#same function performing different actions for different inputs 
def add(x,y):
    return x+y
print(add(3,2)) #addition 
print(add("code" ,"gnan")) #concatenation 
```

```python
#types of polymorphism 
#runtime polymorphism  (method overriding)
#compiletime polymorphism (method overloading)
class person():
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def show(self):
        return f"myself{self.name} and i am a {self.age} years old"
#student 
class student(person):
    def __init__(self,name,age,std):
        #super().__init__(name,age) #no need of self when you use super keyword 
        person.__init__(self,name,age)
        self.std = std

    def show(self):
        return f"myself {self.name} and i am studying {self.std} class"

p= person("rohit",20)
s= student("rohit",20,10)
print(p.show())
print(s.show())

```

```python
#method overloading 
class math:
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
    def add(self , a,b,c,d):
        return a+b+c+d
m = math()
print(m.add(1,2,3,4))
print(m.add(1,2,4))
print(m.add(1,2))
```

```python
#method overloading by using default arguments 
class math:
    def add(self , a,b,c=0,d=0):
        return a+b+c+d
m = math()
print(m.add(1,2,3,4))
print(m.add(1,2,4))
print(m.add(1,2))
```

```python
#method overloading by using default arguments 
class math:
    def add(self , a,b,c=0,d=0,*args):
        print(type(args))
        return a+b+c+d
m = math()
print(m.add(1,2,3,4,5,6,78))
print(m.add(1,2,4))
print(m.add(1,2))
```
