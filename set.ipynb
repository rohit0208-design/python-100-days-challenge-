```python
#union operation  (| symbol)
s1={1,2,3,4}
s2={3,2,14,5,}
s3=s1.union(s2)
print(s3)
```

```python
#intersection operation  (& symbol)
s1={1,2,3,4}
s2={3,2,14,5,}
s3=s1.intersection(s2)
print(s3)
```

```python
#diference operation  (- symbol )
s1={1,2,3,4}
s2={3,2,14,5,}
s3=s1.difference(s2)
print(s3)
```

```python
#symmetric_difference operation 
s1={1,2,3,4}
s2={3,2,14,5,}
s3=s1.symmetric_difference(s2)
print(s3)
```
