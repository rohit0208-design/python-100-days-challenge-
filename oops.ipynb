```python
#  create attribute shares the equal value to all the objects 
class student:    #classname
    school = "DAV school" # classattributes
    def __init__(self,name,age,cls): #intialize constructor (outside constructor they are class attributes)
        self.name = name  # instance attributes
        self.age = age      # they are insisde the constructor 
        self.cls = cls      
       
    def details(self): #method
        return f"My name is {self.name} and {self.age} years old and belong to {self.cls}"
    def percentage(self,percentage):
        return f"Myself {self.name} and my percentage is {percentage}"
s1 = student("rohit", 14,9)
s2 = student("pandu",12,8)
print(s1.details())
print(s2.details())
```
